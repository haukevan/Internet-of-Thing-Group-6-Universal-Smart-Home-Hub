```python
import requests
import json
import numpy as np

x = requests.get('https://io.adafruit.com/api/v2/evanhauk/groups/default/feeds/digital/data/')

y = json.loads(x.text)

#print(y)

print( y[0]['id'] )
print( y[1]['value'] )
print( y[1]['created_at'] )
```

```python
import matplotlib.pyplot as plt

#initialize array of size of json data set for x and y axis
yAxis = np.zeros(len(y))
xAxis = [None] * len(y)

#grab each value from x and y axis
for i in range(100):
  yAxis[i] = y[i]['value']
  xAxis[i] = y[i]['created_at']

#get key name value, knowing what key location they are at
j = 0
for key in y[0].keys():
  if (j == 1):
    yAxisName = key
  elif (j == 4):
    xAxisName = key
  j+=1



```

```python
#plot graph
plt.figure(figsize=[50, 20])
plt.plot(xAxis,yAxis, linewidth=10)
plt.title('Value vs. Time', fontsize=100)
plt.xlabel(xAxisName)
plt.ylabel(yAxisName)
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
plt.xticks(rotation=90)

plt.show()
```
